```python
import pandas as pd
import os
os.getcwd()
centroid = pd.read_csv(os.getcwd()+'/input', sep="\t", header=None)
centroid = centroid[[1,3,4]]
centroid.head()
```

```python
trump = pd.read_csv('3.csv', header=None)
trump.loc[trump[2] == 'Negative', [1]] = trump[1]*-1
trump[0] = trump[0].str.split(',',n=1,expand=True)[0]
trump.head(10)

```

```python
group = trump.groupby(0) \
       .agg({ 1 : ['mean'], 2 : ['size']}) \
       .reset_index()
```

```python
group.head(10)
```

```python
joined = group.set_index(0).join(centroid.set_index(1))
joined = joined.loc[joined[3].notnull()]
```

```python
joined.head(10)
```

```python
joined.to_csv('4.csv')
```
